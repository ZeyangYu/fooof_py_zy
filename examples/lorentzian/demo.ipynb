```python
from fooof.objs import FOOOF, FOOOFGroup
from fooof.plts.aperiodic import plot_aperiodic_params
from fooof.utils.download import load_fooof_data
import numpy as np
from matplotlib import pyplot as plt
```

```python
freq_range = [1, 40]
```

```python
freqs = load_fooof_data('freqs.npy')
spectrum = load_fooof_data('spectrum.npy')

fm_lorentzian = FOOOF( aperiodic_mode = 'lorentzian')
fm_lorentzian.report(freqs, spectrum, freq_range, plt_log=True)
```
